Image Classification using Histogram
=============================================
In this Experiment we will try to classify images or paintings based on the color histogram. 

Feature Extraction
-------------------------
According to the paper [Swain](https://staff.fnwi.uva.nl/r.vandenboomgaard/UvAwiki/uploads/CV0708/swainballard.pdf) histograms are a good set of features for image classification. Also accoring to [Olivier99] it is better to use HSV histogram as it is more invariant to illumination change.

Classification
---------------------

Once we extract the features we will use SVM for classification. The hyperparameters are tuned using cross validation. This paper investigates use of SVM and histogram [Olivier99](http://olivier.chapelle.cc/pub/tnn99.pdf)


First let us load an image and extract its color histogram.

```python
# Loading all the headers
%matplotlib inline
from sklearn.metrics import classification_report,f1_score
import numpy as np
import matplotlib.pyplot as plt
from random import shuffle
import os
import skimage.io as io
from IPython.display import display, Image
from glob import glob
import PIL
from sklearn.cross_validation import train_test_split
import os
import cv2

# ############################# load data###############
data_dir = '../data/two_class_full_size/'
classes = ['color-field-painting', 'realism']
print 'Getting data from ', os.path.abspath(data_dir)
print 'The classes are ', classes
files_list = []  # This will hold all the image models
for genre_i in classes:
    files_list += glob(data_dir + genre_i + '/*.jpg')
# Pick a random image
shuffle(files_list)
sample_images = [PIL.Image.open(f) for f in files_list[:4]]
```

We have selected 4 random images, let us view them and also plot their histogram.

```python
for im in sample_images:
    plt.figure()
    plt.imshow(im)
    plt.title(im.filename)
    plt.figure()
    color = ('b','g','r')
    for channel,col in enumerate(color): 
        histr = cv2.calcHist([np.array(im)],[channel],None,[16],[0,256]) 
        plt.plot(histr,color = col) 
```

Histogram
----------------

Question? 

1 **Bins** How many bins to use, and what dimension? Paper <cite data-cite="chapelle1999support">[chapelle1999support]</cite> says $16^3$ works well. For now I will go with the defaults

Classification
--------------------

**Plan** 

1. Read all the paintings. Extract histogram features.
2. Build a linear classifier using Scikit

**Data**
Each image will have dimension of $3\times binsize$. Need to create a numpy matrix of size dimensionxsamples and corresponding y array

```python
bins = 16
channels = range(3)  # Use all thre RGB
files_class_list = []  # This will hold all the image models
number_of_classes = len(classes)
histograms = []  # a list of histograms in the numpy format where each element is a row
labels = []  # indicates which class it belongs to
for current_class in range(number_of_classes):  # For each genre
    current_genre_file_list = glob(data_dir + classes[current_class] + '/*.jpg')  # List all the filenames
    cv_images = [cv2.imread(f) for f in current_genre_file_list]  # Load the image from filename
    for image_i in cv_images:
            hsv = cv2.cvtColor(image_i,cv2.COLOR_BGR2HSV)
            histograms.append(\
                np.vstack([cv2.calcHist(np.array(hsv), [channel], None,\
                                        [bins],[0,256]) for channel in channels]).T)
            labels.append(current_class)
X = np.vstack(histograms)
# Data processing done, display statistics
print 'The size of data is (samples, dimension) = ', X.shape
```

Classifier
-----------
Let us choose a linear classifier like the one used in [Karayev](http://arxiv.org/abs/1311.3715). **Lasso*** is a linear classifer which ha a good regularization to remove unwanted terms. For now I will use a random value of L1 (alpha = 0.1) later this has to be determined by cross validation

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,labels)


# Linear classifier
from sklearn import linear_model
clf = linear_model.SGDClassifier()

y_pred = clf.fit(X_train, y_train).predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, target_names=classes))
```

Parameter Tuning
--------------------------

We have to tune parameter number of bins, alpha.

First let us tune bins. Let us find the f1 score for different bins

```python
# ############################# load data###############
data_dir = '../data/two_class_full_size/'

classes = ['color-field-painting', 'realism']
print 'Getting data from ', os.path.abspath(data_dir)
print 'The classes are ', classes
files_list = []  # This will hold all the image models
f1_score_list = []  # Hold the f1 scores to plot against bins
for bins in [8,16,32,64,128,256]:
    channels = range(3)  # Use all thre RGB
    files_class_list = []  # This will hold all the image models
    number_of_classes = len(classes)
    histograms = []  # a list of histograms in the numpy format where each
    # element is a row
    labels = []  # indicates which class it belongs to
    for current_class in range(number_of_classes):  # For each genre
        current_genre_file_list = glob(data_dir + classes[current_class]
                                       + '/*.jpg')  # List all the filenames
        cv_images = [cv2.imread(f) for f in current_genre_file_list]
        # Load the image from filename
        for image_i in cv_images:
            hsv = cv2.cvtColor(image_i,cv2.COLOR_BGR2HSV)
            histograms.append(\
                np.vstack([cv2.calcHist(np.array(hsv), [channel], None,\
                                        [bins],[0,256]) for channel in channels]).T)
            labels.append(current_class)
    X = np.vstack(histograms)
    X_train, X_test, y_train, y_test = train_test_split(X,labels)

    # ############################ classification ##################
    # Linear classifier
    from sklearn import linear_model
    clf = linear_model.SGDClassifier()
    y_pred = clf.fit(X_train, y_train).predict(X_test)
    f1_score_list.append(f1_score(y_test, y_pred))
plt.plot([8,16,32,64,128,256],f1_score_list)
plt.title('F1 score vs bins')
plt.xlabel('bins')
plt.ylabel('F1 score')

```

Gridsearch of Parameters
-----------------------------------

Next let us tune the hyperparameter of linear classifier 

```python
from sklearn import datasets
from sklearn.cross_validation import train_test_split
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.svm import SVC
import os 

# ############################# load data###############
data_dir = '../data/two_class_full_size/'
classes = ['color-field-painting', 'realism']
print 'Getting data from ' , os.path.abspath(data_dir)
print 'The classes are ', classes
files_list = []  # This will hold all the image models
f1_score_list = []  # Hold the f1 scores to plot against bins
bins = 16
channels = range(3)  # Use all thre RGB
files_class_list = []  # This will hold all the image models
number_of_classes = len(classes)
histograms = []  # a list of histograms in the numpy format where each
# element is a row
labels = []  # indicates which class it belongs to
for current_class in range(number_of_classes):  # For each genre
    current_genre_file_list = glob(data_dir + classes[current_class]
                                   + '/*.jpg')  # List all the filenames
    cv_images = [cv2.imread(f) for f in current_genre_file_list]
    # Load the image from filename
    for image_i in cv_images:
        hsv = cv2.cvtColor(image_i,cv2.COLOR_BGR2HSV)
        histograms.append(\
                np.vstack([cv2.calcHist(np.array(hsv), [channel], None,\
                                        [bins],[0,256]) for channel in channels]).T)
        labels.append(current_class)
X = np.vstack(histograms)

# Data processing done, display statistics
print 'The size of data is (samples, dimension) = ', X.shape
X_train, X_test, y_train, y_test = train_test_split(X,labels)


# ############################ classification ##################
# Set the parameters by cross-validation
print 'Starting classification'
#tuned_parameters = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-4],
#                     'C': [1, 10, 100, 1000]},
#                    {'kernel': ['linear'], 'C': [1, 10, 100, 1000]}]
tuned_parameters = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-4],
                    'C': [1,10, 100,1000]}]
scores = ['f1']

for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()

    clf = GridSearchCV(SVC(C=1), tuned_parameters, cv=5, scoring=score)
    clf.fit(X_train, y_train)

    print("Best parameters set found on development set:")
    print()
    print(clf.best_estimator_)
    print()
    print("Grid scores on development set:")
    print()
    for params, mean_score, scores in clf.grid_scores_:
        print("%0.3f (+/-%0.03f) for %r"
              % (mean_score, scores.std() / 2, params))
    print()

    print("Detailed classification report:")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test, clf.predict(X_test)
    print(classification_report(y_true, y_pred))
    print()
```

Multi dimensional Histogram
-------------------------------------

So far all the channels were calculated independently. Let us calculate a 3D histogram

```python
# ############################# load data###############
data_dir = '../data/two_class_full_size/'

classes = ['color-field-painting', 'realism']
print 'Getting data from ', os.path.abspath(data_dir)
print 'The classes are ', classes
files_list = []  # This will hold all the image models
f1_score_list = []  # Hold the f1 scores to plot against bins
bins = [16,16,16]
channels = [0,1,2]  # Use all three channels
limits = [0,180,0,256,0,256]  # HSV ranges
files_class_list = []  # This will hold all the image models
number_of_classes = len(classes)
histograms = []  # a list of histograms in the numpy format where each
# element is a row
labels = []  # indicates which class it belongs to
for current_class in range(number_of_classes):  # For each genre
    current_genre_file_list = glob(data_dir + classes[current_class]
                                   + '/*.jpg')  # List all the filenames
    cv_images = [cv2.imread(f) for f in current_genre_file_list]
    # Load the image from filename
    for image_i in cv_images:
        hsv = cv2.cvtColor(image_i,cv2.COLOR_BGR2HSV)
        histograms.append(cv2.calcHist(np.array(hsv), channels, None,\
                                    bins,limits).ravel())
        labels.append(current_class)
X = np.vstack(histograms)
X_train, X_test, y_train, y_test = train_test_split(X,labels)

# ############################ classification ##################
# Linear classifier
from sklearn import linear_model
clf = linear_model.SGDClassifier()
y_pred = clf.fit(X_train, y_train).predict(X_test)
print(classification_report(y_test, y_pred, target_names=classes))
```
