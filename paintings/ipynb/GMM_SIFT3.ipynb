**Step 2**
Load the files for experiment

```python
    # ########## Main ##################
    # since we are just testing BIC we will limit loading image to few images 
    MAX_IMAGE_PER_CLASS = 10  # Load only 10 files at max
    UNIQUE_RUN_NAME = str(uuid.uuid4())  # just a random string
    
    print '-------------------------------------------------------------------'
    print 'Started BIC testing'
    print 'unique run string is ', UNIQUE_RUN_NAME
    data_dir = '..\\data\\two_class_full_size\\'
    classes = [i for i in os.listdir(data_dir) if os.path.isdir(data_dir + i)]
    # classes = ['expressionism', 'realism']
    print 'Getting data from ', os.path.abspath(data_dir)
    print 'The classes are ', classes
    c_image_collection = []  # This will hold all the image models
    for genre_i in classes:
        image_links = glob(data_dir + genre_i + '/*.jpg')
        files = [os.path.abspath(i) for i in image_links]
        image_collection = [misc.imread(i) for i in files]
        image_collection = image_collection[:MAX_IMAGE_PER_CLASS]  # limiter
        for image_i in image_collection:
            curr_GmmImageSIFT = GmmImageSIFT(image_i)
            # set the label for current image
            curr_GmmImageSIFT.label = classes.index(genre_i)
            c_image_collection.append(curr_GmmImageSIFT)
```

BIC Experiment
=========================

Now that we have the data, let us try the BIC experiment for different components and GMM covariance type

```python
X = get_sift_des(sample_image_1.image)
print X.shape
```

Let us check the BIC scores of a single image.

```python
n_components_range = [2**i for i in range(3,7)]+[103,104]  # check for 2,4,8,...128

lowest_bic = np.infty
bic = []
# n_components_range = range(3,4) 
cv_types = ['diag']
for cv_type in cv_types:
    for n_components in n_components_range:
        # Fit a mixture of Gaussians with EM
        gmm = mixture.GMM(n_components=n_components, covariance_type=cv_type)
        gmm.fit(X)
        bic.append(gmm.bic(X))
        if bic[-1] < lowest_bic:
            lowest_bic = bic[-1]
            best_gmm = gmm        
```

```python
# Plot the BIC scores

bic = np.array(bic)
color_iter = itertools.cycle(['k', 'r', 'g', 'b', 'c', 'm', 'y'])
clf = best_gmm
bars = []
spl = plt.subplot(2, 1, 1)
for i, (cv_type, color) in enumerate(zip(cv_types, color_iter)):
    xpos = np.array(range(len(n_components_range))) + .2 * (i - 2)
    bars.append(plt.bar(xpos, bic[i * len(n_components_range):
                                  (i + 1) * len(n_components_range)],
                        width=.2, color=color))
plt.xticks(range(len(n_components_range)))
plt.ylim([bic.min() * 1.01 - .01 * bic.max(), bic.max()])
plt.title('BIC score per model')
xpos = np.mod(bic.argmin(), len(n_components_range)) + .65 +\
    .2 * np.floor(bic.argmin() / len(n_components_range))
plt.text(xpos, bic.min() * 0.97 + .03 * bic.max(), '*', fontsize=14)
spl.set_xlabel('Number of components')
spl.legend([b[0] for b in bars], cv_types)
plt.show()
```
